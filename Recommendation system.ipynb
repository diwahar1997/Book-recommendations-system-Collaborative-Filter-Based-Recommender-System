Popularity Based Recommender System 

```python
#Here we go will the books for which the rating is more the 2.5 and atleast of 250 votes
ratings_with_name = Ratings.merge(books,on='ISBN')
```

```python
ratings_with_name.head(10)
```

```python
ratings_with_name.shape
```

```python
#we are trying to group by based on the book tile because book title can be duplicate but ISBN will be different
ratings_with_name.groupby('Book-Title').count()
```

```python
num_rating_df=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_rating'},inplace=True)
num_rating_df
```

```python
avg_rating_df=ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popular_df=num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df=popular_df[popular_df['num_rating']>=250].sort_values('avg_rating',ascending=False)
```

```python
popular_df=popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')
```

```python
popular_df
```

```python
popular_df=popular_df[['Book-Title','Book-Author','Image-URL-M','num_rating','avg_rating']]
```

```python
popular_df
```

```python
import pickle
pickle.dump(popular_df,open('popular.pkl','wb'))
```

Collaborative Filter Based Recommender System

```python
var=ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
var=var[var].index
var
```

```python
filter_rating=ratings_with_name[ratings_with_name['User-ID'].isin(var)].sort_values('Book-Rating',ascending=True)
filter_rating.head(4)
```

```python
filter_rating.groupby("Book-Title").count()
```

```python
var=filter_rating.groupby("Book-Title").count()['Book-Rating']>=50
var
```

```python
final_rating=filter_rating[filter_rating['Book-Title'].isin(var[var].index)]
```

```python
pt=final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_score=cosine_similarity(pt)

```

```python
similarity_score
```

```python
def book(name):
    #index
    index =np.where(pt.index==name)[0][0]
    distance=sorted(enumerate(list(similarity_score[0])),key=lambda x:x[1],reverse=True)[0:6]
    for i in distance:
        print(pt.index[i[0]])
    
        
```
